**Business Case:**

Mental Health Insights for Workplace Wellness Initiatives
In recent years, organizations globally have started recognizing the importance of employee mental health. However, identifying how factors like company benefits, remote work options, family history, company size, and industry sector (like tech companies) influence mental health treatment-seeking behavior remains underexplored.

This project aims to bridge that gap by analyzing survey data related to mental health in the workplace, providing actionable insights for businesses to:

Understand how company-provided mental health benefits influence treatment uptake

Identify whether remote work reduces workplace mental health issues

Measure country-wise treatment rates to shape localized employee wellness policies

Evaluate the impact of company culture (like in tech companies) on employees’ mental health decisions

Discover the relationship between family history and treatment-seeking behavior


**Why It Matters for Business:**

Improve employee retention and productivity by addressing mental health proactively

Design better wellness programs and leave policies based on data-driven evidence

Strengthen employer branding and employee satisfaction

Reduce healthcare costs through early intervention and support programs

Shape country-specific or remote work mental health strategies


**What the Analysis Covers:**

Comparative analysis of mental health treatment uptake across various factors

Country-wise and company-type breakdowns

Remote work influence on mental health care access

Benefit programs’ effectiveness in promoting treatment

Family history correlation with mental health management


**Business Outcome**:

By understanding these patterns, businesses can tailor their HR, benefits, and work-from-home policies to create healthier, more supportive workplaces — resulting in improved employee wellbeing, higher job satisfaction, and ultimately better organizational performance.

# **Treatment** **Seeking Behavior Based on Family History**

**Reason : To analyze how family medical history influences employees' decisions to seek treatment.**

```python
grouped_data = data.groupby(['family_history', 'treatment']).size().reset_index(name='count')
steel_gradient = sns.color_palette(["#4682B4", "#B0C4DE"])  # Steel Blue to Light Steel Blue
plt.figure(figsize=(8,6))
sns.barplot(x='family_history', y='count', hue='treatment', data=grouped_data, palette=steel_gradient)
plt.title('Treatment Rates by Family History of Mental Illness', fontsize=14)
plt.xlabel('Family History of Mental Illness', fontsize=12)
plt.ylabel('Number of Responses', fontsize=12)
plt.legend(title='Sought Treatment')
plt.show()

```

# **Mental Health Consequences by Tech Company Employment Status**

**Reason : To examine how working in a tech company relates to experiencing mental health consequences.**

```python
heatmap_data = data.groupby(['tech_company', 'mental_health_consequence']).size().unstack().fillna(0)
plt.figure(figsize=(8,5))
sns.heatmap(heatmap_data, annot=True, fmt="g", cmap="Blues")

plt.title('Mental Health Consequences by Tech Company Status')
plt.xlabel('Mental Health Consequence')
plt.ylabel('Works in a Tech Company?')
plt.show()
```

# **Proportion of Employees Receiving Mental vs Physical Health Treatment by Benefit Availability**

**Reason : To show how the availability of mental health benefits from companies influences whether employees receive mental or physical health treatment.**

```python
filtered_data = data[(data['benefits'].isin(['Yes', 'No'])) & (data['mental_vs_physical'].isin(['Yes', 'No']))]
pivot_data = filtered_data.groupby(['benefits', 'mental_vs_physical']).size().unstack().fillna(0)
percent_data = pivot_data.div(pivot_data.sum(axis=1), axis=0) * 100
percent_data.plot(kind='bar', stacked=True, figsize=(8,6), color=['#4682B4', '#87CEFA'])
plt.title('Proportion of Mental vs Physical Health Treatment by Benefits (Yes/No)')
plt.xlabel('Company Offers Mental Health Benefits?')
plt.ylabel('Percentage of Responses')
plt.legend(title='Mental vs Physical Health Treatment')
plt.xticks(rotation=0)
plt.show()
```

# **Impact of Remote Work Option on Seeking Treatment**

**Reason : To analyze how allowing remote work affects the number of employees who seek or do not seek treatment.**

```python
filtered_data = data[(data['remote_work'].isin(['Yes', 'No'])) & (data['treatment'].isin(['Yes', 'No']))]
grouped_data = filtered_data.groupby(['remote_work', 'treatment']).size().reset_index(name='count')

# Prepare data for each remote_work category
yes_data = grouped_data[grouped_data['remote_work'] == 'Yes'].set_index('treatment')['count']
no_data = grouped_data[grouped_data['remote_work'] == 'No'].set_index('treatment')['count']

colors = ['#4682B4', '#87CEFA']

fig, axs = plt.subplots(1, 2, figsize=(12,6))

# Donut chart for Remote Work = Yes
axs[0].pie(yes_data, labels=yes_data.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.4))
axs[0].set_title('Treatment Status (Remote Work: Yes)')

# Donut chart for Remote Work = No
axs[1].pie(no_data, labels=no_data.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.4))
axs[1].set_title('Treatment Status (Remote Work: No)')

plt.tight_layout()
plt.show()
```

# **Leave Policy Availability by Benefits Provided**

**Reason : To show how different employee benefits are associated with the availability of leave policies.**

```python
import plotly.express as px
import pandas as pd

filtered_data = data[(data['leave'] != "Don't know") & (data['benefits'] != "Don't know")]
grouped_data = filtered_data.groupby(['benefits', 'leave']).size().reset_index(name='Count')

# Aggregate counts for each 'benefits' type, total counts to show funnel base
total_counts = grouped_data.groupby('benefits')['Count'].sum().reset_index()

# Sort benefits by total counts descending for funnel order
total_counts = total_counts.sort_values('Count', ascending=False)

# Prepare data for funnel: sum by leave difficulty for each benefit, to visualize funnel stages
# For funnel, let's focus on leave difficulty categories stacked as stages per benefit
# Here we plot each 'leave' category as a separate funnel segment within each 'benefit'

# Plot funnel chart using plotly.express
fig = px.funnel(
    grouped_data,
    x='Count',
    y='benefits',
    color='leave',
    color_discrete_map={
        'Very easy': '#0077b6',
        'Somewhat easy': '#00b4d8',
        'Somewhat difficult': '#90e0ef',
        'Very difficult': '#caf0f8'
    },
    title="Leave Policy Availability by Benefits Provided (Funnel Chart)"
)

fig.show()
```

# **Leave Difficulty Based on Remote Work**

**Reason : To explore how remote work and work interference affect the ease or difficulty of taking leave.**

```python
import numpy as np

# Filter data
filtered_data = data[
    (data['remote_work'] != "Don't know") &
    (data['leave'].isin(['Very easy', 'Very difficult']))
]

# Pivot table grouped only by remote_work
pivot = filtered_data.pivot_table(
    index='remote_work',
    columns='leave',
    aggfunc='size',
    fill_value=0
).reset_index()

# Sort just to keep order consistent
pivot = pivot.sort_values('remote_work')

labels = pivot['remote_work'].tolist()
x = np.arange(len(labels))

very_easy = pivot['Very easy']
very_difficult = pivot['Very difficult']

plt.figure(figsize=(8,6))

plt.fill_between(x, 0, very_easy, color='#00b4d8', alpha=0.7, label='Very Easy')
plt.fill_between(x, very_easy, very_easy + very_difficult, color='#0077b6', alpha=0.7, label='Very Difficult')

plt.xticks(x, labels)
plt.title('Leave Difficulty by Remote Work Status')
plt.xlabel('Remote Work Status')
plt.ylabel('Number of People')
plt.legend(title='Leave Difficulty')

plt.tight_layout()
plt.show()
```

```python
filtered_data = data[(data['Age'].notna()) & (data['treatment'].notna())]
filtered_data['Age'] = pd.to_numeric(filtered_data['Age'], errors='coerce')
filtered_data = filtered_data[(filtered_data['Age'] >= 18) & (filtered_data['Age'] <= 100)]

# Create age bins
bins = [18, 25, 30, 35, 40, 45, 50, 60, 70, 100]
labels = ['18-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-60', '61-70', '71+']
filtered_data['Age_Group'] = pd.cut(filtered_data['Age'], bins=bins, labels=labels)

# Group by age group and treatment
grouped = filtered_data.groupby(['Age_Group', 'treatment']).size().reset_index(name='Count')

# Line chart using seaborn
plt.figure(figsize=(10,6))
sns.lineplot(data=grouped, x='Age_Group', y='Count', hue='treatment', marker='o', palette='Blues')

plt.title('Age Group vs Treatment Taken')
plt.xlabel('Age Group')
plt.ylabel('Number of People')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
